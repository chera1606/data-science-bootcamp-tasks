# Student Dataset Creation using Pandas

In this task, I created a dataset using a Python dictionary and converted it into a pandas DataFrame.

The dataset contains information about 15 students, including their names, scores, attendance, and status.

## Objectives
- Create a dataset with 5 columns and 15 rows
- Assign a custom index
- Handle missing values properly
- Display the dataset

## Steps Performed

1. Created a dictionary containing student data
2. Assigned a custom index (UGR-101 to UGR-115)
3. Converted the dictionary into a pandas DataFrame
4. Handled missing values using the mean
5. Displayed the cleaned dataset

```python
import pandas as pd

student_data = {
    'Student_Name': [
        'Alex', 'Blake', 'Charlie', 'Dana', 'Eli',
        'Farah', 'Gabe', 'Hope', 'Isaac', 'Jade',
        'Kai', 'Lana', 'Milo', 'Nora', 'Otis'
    ],
    'Math_Score': [
        85, 92, None, 78, 88,
        95, 70, 82, 91, None,
        84, 76, 89, 93, 80
    ],
    'History_Score': [
        90, 85, 88, None, 92,
        80, 75, 85, 88, 91,
        82, 79, 90, 85, 87
    ],
    'Attendance_Rate': [
        0.95, 0.98, 0.92, 0.85, 1.0,
        0.96, 0.88, 0.94, 0.91, 0.99,
        0.87, 0.93, 0.97, 0.90, 0.82
    ],
    'Status': [
        'Pass', 'Pass', 'Pending', 'Pass', 'Pass',
        'Pass', 'Fail', 'Pass', 'Pass', 'Pending',
        'Pass', 'Pass', 'Pass', 'Pass', 'Fail'
    ]
}

student_ids = [f'UGR-{i}' for i in range(101, 116)]

df = pd.DataFrame(student_data, index=student_ids)

df['Math_Score'] = df['Math_Score'].fillna(df['Math_Score'].mean())
df['History_Score'] = df['History_Score'].fillna(df['History_Score'].mean())

print("Missing values after cleaning:")
print(df.isnull().sum())

print("\n--- Student Gradebook Summary ---")
print(df)
```
